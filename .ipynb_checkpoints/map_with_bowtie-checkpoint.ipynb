```python
from pandas.io.parsers import *
import numpy as np
import mysql.connector
import pandas as pd
from pandas.io.sql import *

df = pd.read_table('R1_FBA.sam', header=None, names=['read_name','flags','sgRNA','offset','qual','CIGAR',
                                                    'na','na2','na3','read_seq','read_qual',
                                                    'as','xs','xm','x0','xg','nm','md','yt'])
```

```python
mapped = df[df.flags == 0]
for col in ['as','xs','xm','x0','xg','nm']:
    temp = mapped[col].apply(lambda x:int(x.split(':')[-1]),1)
    mapped[col] = temp
```
